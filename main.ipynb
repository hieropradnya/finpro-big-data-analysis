**2. Produk Terlaris dan Tidak Terlaris**

```python
#klasifikasi data berdasarkan nama barang dan jumlah
produk_terlaris = df.groupby('nama_produk')['kuantitas'].sum().sort_values(ascending=False)

#mengambil 10 produk mie yang paling laris dan paling tidak laris
terlaris = produk_terlaris.head(10)
tidak_laris = produk_terlaris.tail(10)

# tampilkan 10 produk terlaris
print("10 Produk Mie Terlaris:")
print(terlaris)

# tampilkan 10 produk paling tidak laris
print("\n10 Produk Mie Tidak Laris:")
print(tidak_laris)
```

**3. Distribusi Jumlah Pembelian**

```python
#buat visualisasi distribusi jumlah pembelian
sns.histplot(df['kuantitas'], bins=20)
plt.title('Distribusi Jumlah Pembelian')
plt.ylabel('Total Transaksi')
plt.xlabel('Jumlah Pembelian')
plt.show()
```

**4. Tren Penjualan**

```python
# Analisa tren penjualan dalam 1 bulan
# convert kolom tanggal_transaksi menjadi tipe datetime
df['tanggal_transaksi'] = pd.to_datetime(df['tanggal_transaksi'], format='%d/%m/%Y')

# kelompokkan data berdasarkan tanggal dan menghitung total penjualan per hari
penjualan_per_tanggal = df.groupby('tanggal_transaksi')['kuantitas'].sum()



# Buat visualisasinya tren penjualan
plt.figure(figsize=(18, 6))
plt.plot(penjualan_per_tanggal.index, penjualan_per_tanggal.values)
plt.title('Tren Penjualan November 2024')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Produk Terjual')
plt.show()
```

**5. Pendapatan Kotor yang Didapatkan**

```python
# Omset harian
penjualan_per_tanggal = df.groupby('tanggal_transaksi')['total'].sum()
print(penjualan_per_tanggal)
```

```python
# analisa omset dalam 1 bulan
# convert kolom tanggal_transaksi menjadi tipe datetime
df['tanggal_transaksi'] = pd.to_datetime(df['tanggal_transaksi'], format='%d/%m/%Y')

# kelompokkan data berdasarkan tanggal dan menghitung total penjualan per hari
penjualan_per_tanggal = df.groupby('tanggal_transaksi')['total'].sum()




# visualisasi omset dalam sebualan
plt.figure(figsize=(18, 6))
plt.plot(penjualan_per_tanggal.index, penjualan_per_tanggal.values)
plt.title('Omset November 2024')
plt.xlabel('Tanggal')
plt.ylabel('Pendapatan Kotor')
plt.show()
```

**6. Mengelompokkan Jenis Pelanggan**

```python
#mengelompokkan pelanggan berdasarkan frekuensi pembelian
frekuensi_pembelian = df.groupby('nama_pelanggan')['nama_produk'].count()

#kategori kondisi
def segmentasi_pelanggan(x):
    if x > 10:
        return 'Pelanggan Loyal'
    elif x > 1:
        return 'Pelanggan Biasa'
    else:
        return 'Pelanggan Baru'

# menambahkan kolom segmentasi pada data
df['segmentasi'] = df['nama_pelanggan'].map(frekuensi_pembelian).map(segmentasi_pelanggan)

# tampilkan hasil
print(df[['nama_pelanggan', 'segmentasi']])
```

```python
# Menghitung jumlah segmen berdasarkan kolom 'segmentasi'
segmen_counts = df['segmentasi'].value_counts()

# Membuat bar chart
fig, ax = plt.subplots()
ax.bar(segmen_counts.index, segmen_counts.values, color=['blue', 'orange', 'green'])
ax.set_title('Jumlah Pembelian Berdasarkan Segmen')
ax.set_xlabel('Segmen Pelanggan')
ax.set_ylabel('Jumlah Pembelian')

# Menampilkan angka di atas bar
for i, v in enumerate(segmen_counts.values):
    ax.text(i, v + 0.5, str(v), ha='center')

plt.show()
```

**7. Pelanggan dengan Jumlah Pembelian Terbanyak dan Tersedikit**

```python
# Mengelompokkan pelanggan berdasarkan jumlah pembelian
frekuensi_pembelian = df.groupby('nama_pelanggan')['nama_produk'].count()

# Menampilkan 5 pelanggan dengan pembelian terbanyak
top_5_pelanggan_terbanyak = frekuensi_pembelian.sort_values(ascending=False).head(5)

# Menampilkan 5 pelanggan dengan pembelian tersedikit
top_5_pelanggan_tersedikit = frekuensi_pembelian.sort_values(ascending=True).head(5)

# Menampilkan hasil
print("5 Pelanggan dengan Pembelian Terbanyak:")
print(top_5_pelanggan_terbanyak)

print("\n5 Pelanggan dengan Pembelian Tersedikit:")
print(top_5_pelanggan_tersedikit)
```
